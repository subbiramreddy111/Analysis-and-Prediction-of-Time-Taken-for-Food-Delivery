### Statistical analysis

```python
from scipy.stats import shapiro,levene
from scipy.stats import f_oneway
```

```python
shapiro(new_df["Time_taken (min)"])
```

### Hypothesis 1: Impact of City on time taken for delivery

```python
#Creating individual group based on City type and their respective time taken for delivery

Metropolitian_del_time=new_df[new_df["City"]=="Metropolitian"]["Time_taken (min)"]
Urban_del_time=new_df[new_df["City"]=="Urban"]["Time_taken (min)"]
Semi_Urban_del_time=new_df[new_df["City"]=="Semi-Urban"]["Time_taken (min)"]

#Using Anova test
#Hypothesis 
#Ho delivery time significantly not differ with type of city
#Ha delivery time significantly differ with type of city
#alpha- 0.05

print(f_oneway(Metropolitian_del_time,Urban_del_time,Semi_Urban_del_time))
```

```python
#pvalue is 0.0 which is less than 0.05 and we are rejecting null hypothesis.
#so we can infer delivery time differ with type of city.
```

### Hypothesis 2: Impact of Weather condition on time taken for delivery

```python
#Creating individual group based on Weather conditions and their respective time taken for delivery

Fog_del_time=new_df[new_df["Weather conditions"]=="Fog"]["Time_taken (min)"]
Stromy_del_time=new_df[new_df["Weather conditions"]=="Stormy"]["Time_taken (min)"]
Cloudy_del_time=new_df[new_df["Weather conditions"]=="Cloudy"]["Time_taken (min)"]
Sandstorms_del_time=new_df[new_df["Weather conditions"]=="Sandstorms"]["Time_taken (min)"]
Windy_del_time=new_df[new_df["Weather conditions"]=="Windy"]["Time_taken (min)"]
Sunny_del_time=new_df[new_df["Weather conditions"]=="Sunny"]["Time_taken (min)"]

#Using Anova test
#Hypothesis 
#Ho delivery time significantly not differ with type of weather condition
#Ha delivery time significantly differ with type of weather condition
#alpha- 0.05

print(f_oneway(Fog_del_time,Stromy_del_time,Cloudy_del_time,Sandstorms_del_time,Windy_del_time,Sunny_del_time))
```

```python
#pvalue is 0.0 which is less than 0.05 and we are rejecting null hypothesis.
#so we can infer delivery time differ with type of weather condition.
```

### Hypothesis 3: Impact of Type of Order on time taken for delivery

```python
#Creating individual group based on type of order and their respective time taken for delivery.

Snack_del_time=new_df[new_df["Type_of_order"]=="Snack"]["Time_taken (min)"]
Meal_del_time=new_df[new_df["Type_of_order"]=="Meal"]["Time_taken (min)"]
Drinks_del_time=new_df[new_df["Type_of_order"]=="Drinks"]["Time_taken (min)"]
Buffet_del_time=new_df[new_df["Type_of_order"]=="Buffet"]["Time_taken (min)"]

#Using Anova test
#Hypothesis 
#Ho delivery time significantly not differ with type of order
#Ha delivery time significantly differ with type of order
#alpha- 0.05

print(f_oneway(Snack_del_time,Meal_del_time,Drinks_del_time,Buffet_del_time))
```

```python
#pvalue is 0.377 which is greater than 0.05 and we fail to reject null hypothesis.
#so we can infer delivery time does not differ with the type of order.
```

```python
perf_score = pd.DataFrame(columns=['Model', 'Alpha', 'L1_Ratio','R2_Train','R2_Test','RMSE Train','RMSE TEST','MAPE' ] )
```

```python
def per_measures(model):

   #-----------------------------------------------#
   # Calculation of R square and Adj R2            #
   #-----------------------------------------------#

   # calculate rsquare value

    rsq_train = model.score(xtrain, ytrain)
    rsq_test  = model.score(xtest, ytest)

    # calculating adjusted rsquare value
    n =xtrain.shape[0]
    k = xtrain.shape[1]
    rsq_adj = 1 - ((1-rsq_test)*(n-1)/(n-k-1))


   #-----------------------------------------------#
   # Model getting trained                         #
   #-----------------------------------------------#

    model.fit(xtrain,ytrain)

    pred_train = model.predict(xtrain)
    pred_test = model.predict(xtest)

   #-----------------------------------------------#
   # Calculation of RMSE and MAPE                  #
   #-----------------------------------------------#

    rmse_train =np.sqrt(mean_squared_error(ytrain,pred_train))
    rmse_test  =np.sqrt(mean_squared_error(ytest,pred_test))


    MAPE_test = (np.mean(np.abs((ytest - pred_test) / ytest)) * 100)


    return (rsq_train,rsq_test,rmse_train,rmse_test,MAPE_test)
```

```python
def update_performance (name, model, alpha = '-', l1_ratio = '-' ):


    # assign 'comp_perf' as global variable
    global perf_score


    # append the results to the dataframe 'score_card'
    # 'ignore_index = True' do not consider the index labels
    perf_score = perf_score.append({'Model'          : name,
                                    'Alpha'          : alpha,
                                    'L1_Ratio'       : l1_ratio,
                                    'R2_Train'       : per_measures(model)[0],
                                    'R2_Test'        : per_measures(model)[1],
                                    'RMSE Train'     : per_measures(model)[2],
                                    'RMSE TEST'      : per_measures(model)[3],
                                    'MAPE'           : per_measures(model)[4]
                                    },
                                   ignore_index = True)
```
